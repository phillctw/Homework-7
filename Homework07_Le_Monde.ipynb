```python
from bs4 import BeautifulSoup
import requests
```

```python
response = requests.get('https://www.lemonde.fr')
doc = BeautifulSoup(response.text)
```

```python
doc.find_all('div', attrs={'class':'article__title'})
```

```python
items = doc.find_all(class_='article')
# len(items)
```

```python
le_monde = [ ]
for item in items:
    # print('---------')
    l_mo = { }
    l_mo['title'] = item.find(class_='article__title').text
    try:
        l_mo['subhead'] = item.find(class_='article__desc').text
    except:
        pass
    try:
        l_mo['URL'] = item.find('a').get('href')
    except:
        pass
    try:
        l_mo['premium'] = item.find(class_='icon__premium').text
    except:
        pass
    try:
        l_mo['byline'] = item.find(class_='article__byline').text
    except:
        pass
    try:
        l_mo['article type'] = item.find(class_='article__type').text
    except:
        pass
    try:
        l_mo['image URL'] = item.find('img').get('data-src')
    except:
        pass
    # print(l_mo)
    le_monde.append(l_mo)
```

```python
le_monde
```

```python
import pandas as pd
df = pd.DataFrame(le_monde)
```

```python
df.to_csv('le_monde.csv', index=False)
```
